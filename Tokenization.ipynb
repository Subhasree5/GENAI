```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize
```

```python
text = "HII,Myself Subhasree Ragu. I am CSE Student."
sentences = sent_tokenize(text)
print(sentences)
```

```python
from nltk.tokenize import word_tokenize
```

```python
sentence= "HII,Myself Subhasree Ragu. I am CSE Student."
words= word_tokenize(sentence)
print(words)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
tokens = tokenizer.tokenize("Have a look at NLTK's tokenizers.")
print(tokens)
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(
    "Custom rule: keep only words & numbers, drop punctuation!")
print(tokens)
```
