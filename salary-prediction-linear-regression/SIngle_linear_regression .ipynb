# **Simple Linear Regression - Salary Prediction**

#**Objective:To predict salary based on years of experience using Linear Regression.**

```python
import pandas as pd
```

```python
data=pd.read_csv("/content/Salary_Data.csv")
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['YearsExperience'],data['Salary'])
```


## **We use Linear Regression from sklearn to train the model.**

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(data[['YearsExperience']],data['Salary'])
```

# **Predict salary for 5 years of experience.**

```python
model.predict([[5]])
```
